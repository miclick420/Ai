```python
import pandas 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Your Burger King data for Yelp Reviews and Google Reviews
data = {
    'Yelp Reviews': [240, 150, 120, 190, 140, 110, 200, 100, 130, 160, 210, 115, 135, 185, 95, 175, 145, 220, 125, 85],
    'Google Reviews': [310, 280, 270, 330, 250, 220, 310, 210, 260, 280, 320, 230, 240, 300, 200, 290, 270, 340, 230, 180]
}

df = pd.DataFrame(data)

# Calculate Pearson correlation coefficient (r) using numpy
correlation = np.corrcoef(df['Yelp Reviews'], df['Google Reviews'])[0, 1]

# Calculate coefficient of determination (r squared)
r_squared = correlation ** 2

print(f"Correlation coefficient (r): {correlation:.2f}")
print(f"Coefficient of determination (r^2): {r_squared:.2f}")

# Create scatter plot with regression line
sns.regplot(x='Yelp Reviews', y='Google Reviews', data=df, scatter=True, line_kws={"color": "red"})

# Add title and labels
plt.title('Burger King: Yelp Reviews vs Google Reviews')
plt.xlabel('Yelp Reviews')
plt.ylabel('Google Reviews')

# Add r and r^2 on the plot
plt.text(
    x=df['Yelp Reviews'].min(),
    y=df['Google Reviews'].max() * 0.9,
    s=f"r = {correlation:.2f}\nr = {r_squared:.2f}",
    fontsize=12,
    color='orange'
)

plt.grid(True)
plt.show()
```

```python
import pandas as pd
from io import StringIO

# Load your clean data
csv_data = '''
Name,City,Yelp Rating,Yelp Reviews,Google Rating,Google Reviews
Burger King,Denver,3.8,240,3.9,310
Burger King,Aurora,3.5,150,3.7,280
Burger King,Fort Collins,4.0,120,4.1,270
Burger King,Colorado Springs,3.7,190,3.8,330
Burger King,Lakewood,3.6,140,3.7,250
Burger King,Thornton,3.4,110,3.5,220
Burger King,Arvada,3.8,200,3.9,310
Burger King,Westminster,3.3,100,3.6,210
Burger King,Pueblo,3.7,130,3.8,260
Burger King,Centennial,3.5,160,3.7,280
Burger King,Broomfield,3.9,210,4.0,320
Burger King,Longmont,3.6,115,3.7,230
Burger King,Castle Rock,3.5,135,3.6,240
Burger King,Douglas County,3.8,185,3.9,300
Burger King,Englewood,3.4,95,3.5,200
Burger King,Greeley,3.7,175,3.8,290
Burger King,Littleton,3.6,145,3.7,270
Burger King,Highlands Ranch,3.9,220,4.1,340
Burger King,Commerce City,3.5,125,3.6,230
Burger King,Northglenn,3.3,85,3.4,180
'''

# Read the CSV into a DataFrame
df = pd.read_csv(StringIO(csv_data))

# Function to detect outliers using IQR
def detect_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Detect outliers
yelp_outliers = detect_outliers('Yelp Reviews')
google_outliers = detect_outliers('Google Reviews')

print(" Yelp Review Outliers:")
print(yelp_outliers[['City', 'Yelp Reviews']])

print("\n Google Review Outliers:")
print(google_outliers[['City', 'Google Reviews']])
```
