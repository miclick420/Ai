How does the distribution of Speed stats compare between first-generation Legendary Pokemon and non-Legendary Pokémon, and which group has the higher average Speed?

To answer this question, I would need to: Create two histograms
1: one for the Speed stats of Legendary Pokemon and one for non-Legendary Pokemon.
2: Calculate measures of central tendency, specifically the mean, median, and possibly the mode for each group.
3: Compare the distributions and averages to determine which group tends to be faster and how their speed values are spread out. This encourages both graphical analysis (via histograms) and numerical summary (via central tendency), as well as a clear comparison. 

Legendary Pokémon in Generation 1 tend to have higher Speed stats on average than non-Legendaries, and their distribution is more concentrated toward the upper end of the Speed spectrum.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Try to load your file if it exists
try:
    df = pd.read_csv('pokemon_gen1.csv')
except FileNotFoundError:
    # If file doesn't exist, create a sample dataset for demonstration
    print("Creating sample data since 'pokemon_gen1.csv' was not found")
    
    import numpy as np
    np.random.seed(42)
    n = 150  # Number of Pokémon in Gen 1
    data = {
        'name': [f'Pokemon{i}' for i in range(1, n+1)],
        'generation': [1] * n,
        'speed': np.random.normal(70, 25, n).astype(int),
        'is_legendary': [0] * (n - 10) + [1] * 10  # 10 legendary Pokémon
    }
    df = pd.DataFrame(data)

# Filter Generation 1 Pokémon
gen1 = df[df['generation'] == 1]

# Create histograms of Speed stats for Legendary vs Non-Legendary
plt.figure(figsize=(10, 6))
sns.histplot(data=gen1, x='speed', hue='is_legendary', bins=15, kde=True, palette={0: "blue", 1: "red"})
plt.title('Speed Distribution: Legendary vs Non-Legendary Pokémon (Gen 1)')
plt.xlabel('Speed')
plt.ylabel('Number of Pokémon')
plt.legend(title='Is Legendary', labels=['Non-Legendary', 'Legendary'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

The histogram shows that Legendary Pokemon generally have higher Speed stats than Non-Legendary ones. Most Legendary Pokemon are grouped on the higher end of the Speed range, often between 90 and 150, while Non-Legendary Pokemon are spread more widely, with many falling between 50 and 70. Although there are far fewer Legendary Pokemon, their speeds are consistently higher, indicating they’re designed to be faster overall. Non Legendaries vary more, including both slow and fast Pokemon, but average lower in Speed compared to Legendaries.

Legendary Pokémon in Generation 1 are typically faster than Non-Legendary Pokémon, and the histogram clearly shows this difference in distribution.

```python
# Mean (average) Speed
mean_speed_legendary = gen1[gen1['is_legendary'] == 1]['speed'].mean()
mean_speed_nonlegendary = gen1[gen1['is_legendary'] == 0]['speed'].mean()

# Median Speed
median_speed_legendary = gen1[gen1['is_legendary'] == 1]['speed'].median()
median_speed_nonlegendary = gen1[gen1['is_legendary'] == 0]['speed'].median()

# Mode Speed
mode_speed_legendary = gen1[gen1['is_legendary'] == 1]['speed'].mode()[0]
mode_speed_nonlegendary = gen1[gen1['is_legendary'] == 0]['speed'].mode()[0]

print(f"Legendary Pokémon - Mean: {mean_speed_legendary:.2f}, Median: {median_speed_legendary}, Mode: {mode_speed_legendary}")
print(f"Non-Legendary Pokémon - Mean: {mean_speed_nonlegendary:.2f}, Median: {median_speed_nonlegendary}, Mode: {mode_speed_nonlegendary}")
```
