```python
import random
import seaborn as sns
import matplotlib.pyplot as plt

# Seaborn style
sns.set(style="whitegrid")

# Function to roll and plot
def roll_and_plot(n_sides=6, n_trials=100):
    # Simulate die rolls
    rolls = [random.randint(1, n_sides) for _ in range(n_trials)]

    # Plot
    sns.histplot(rolls, bins=n_sides, discrete=True)
    plt.title(f"{n_sides}-Sided Die: {n_trials} Rolls")
    plt.xlabel("Die Face")
    plt.ylabel("Frequency")
    plt.show()
```

```python
roll_and_plot(n_sides=6)   # For a 6-sided die
```

```python
sns.set(style="whitegrid")
# Plot for 6-sided die
sns.histplot(rolls_6, bins=sides_6, discrete=True)
plt.title(f"6-Sided Die: {trials_6} Rolls")
plt.xlabel("Die Face")
plt.ylabel("Frequency")
plt.show()
```

```python
roll_and_plot(n_sides=12)  # For a 12-sided die
```

```python
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # 3D plotting
import numpy as np

# Function to roll dice and make 3D bar plot
def roll_and_plot_3d(n_sides=6, n_trials=100):
    # Simulate die rolls
    rolls = [random.randint(1, n_sides) for _ in range(n_trials)]

    # Count frequencies
    faces = list(range(1, n_sides + 1))
    frequencies = [rolls.count(face) for face in faces]

    # 3D plot setup
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')

    # Positioning bars
    xpos = np.array(faces)           # X: Die faces
    ypos = np.zeros_like(xpos)       # Y: All bars at same Y
    zpos = np.zeros_like(xpos)       # Z: Bars rise from 0

    dx = np.ones_like(xpos) * 0.5    # Width of bars
    dy = np.ones_like(ypos) * 0.5    # Depth of bars
    dz = np.array(frequencies)       # Height = frequency

    # Create 3D bars
    ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color='skyblue', edgecolor='black')

    # Labeling
    ax.set_xlabel('Die Face')
    ax.set_ylabel('Category (Fixed)')
    ax.set_zlabel('Frequency')
    ax.set_title(f"{n_sides}-Sided Die: {n_trials} Rolls (3D)")

    # Fix X ticks
    ax.set_xticks(faces)
    ax.set_yticks([])  # Hide Y-axis categories

    plt.show()
```

```python
roll_and_plot_3d(n_sides=6)
```

```python
roll_and_plot_3d(n_sides=12)


```
